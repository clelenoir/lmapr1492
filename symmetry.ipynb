```python
from pymatgen.ext.matproj import MPRester
from pymatgen.io.cif import CifWriter
from pymatgen.symmetry.analyzer import *
from pymatgen.io.cif import *
from pymatgen.core.operations import *
from pymatgen.symmetry.site_symmetries import *
import matplotlib.pyplot as plt
import matplotlib.image as img
```

```python
with MPRester("inU5O5NjJKENIfaZf6EO") as m:
    structure = m.get_structure_by_material_id("mp-20289")
    #téléchargement fichier cif
    w = CifWriter(structure) 
    w.write_file('NaInS2.cif')
    
    #Trouver les symétries
    #s = SpacegroupAnalyzer(structure)
    #print(s.get_symmetry_operations())
    
    #Première opération de symétrie
    Sym1 = SymmOp.from_xyz_string('-x,-y,-z')#2
    Na0 = [0.,0.,0.5]#1
    Na0S = Sym1.operate(Na0)
    print("-------------------------------")
    print("Première opération de symétrie: \n")
    print("atome de départ :", Na0)
    print("operation de symetrie : \n", Na0S, '=', Sym1.translation_vector, "+ \n", Sym1.rotation_matrix, ".", Na0)
    print("atome d'arrivé :", Na0S)
    print()
    Image1 = img.imread('symétrie1.PNG')
    fig,axes = plt.subplots(figsize=(10,10))
    axes.get_xaxis().set_visible(False)
    axes.get_yaxis().set_visible(False)
    axes.imshow(Image1)
    plt.show()
    
```

```python
print("-------------------------------")
print("Deuxième opération de symétrie: \n")
Sym2 = SymmOp.from_xyz_string('-y,-x,z')#8 plan miroir
In1 = [1/3,2/3,2/3] #12
In1S = Sym2.operate(In1)
print("atome de départ :", In1)
print("operation de symetrie : \n", In1S, '=', Sym2.translation_vector, "+ \n", Sym2.rotation_matrix, ".", In1)
print("atome d'arrivé :", In1S)
print()
Image2 = img.imread('symétrie2.PNG')
fig,axes = plt.subplots(figsize=(10,10))
axes.get_xaxis().set_visible(False)
axes.get_yaxis().set_visible(False)
axes.imshow(Image2)
plt.show()
    
```

```python
#Troisième opération de symétrie
print("-------------------------------")
print("Troisième opération de symétrie: \n")
Sym3 = SymmOp.from_xyz_string('x-1/3,y+1/3,z+1/3')#13 translation
S2 = [2/3,1/3,(0.259876+1/3)] #8
S2S = Sym3.operate(S2)
print("atome de départ :", S2)
print("operation de symetrie : \n", S2S, '=', Sym3.translation_vector, "+ \n", Sym3.rotation_matrix, ".", S2)
print("atome d'arrivé :", S2S)
print()
Image3 = img.imread('symétrie3.PNG')
fig,axes = plt.subplots(figsize=(10,10))
axes.get_xaxis().set_visible(False)
axes.get_yaxis().set_visible(False)
axes.imshow(Image3)
plt.show()
```
