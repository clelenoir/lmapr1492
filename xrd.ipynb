```python
from pymatgen.ext.matproj import MPRester
from pymatgen.analysis.diffraction.xrd import XRDCalculator 
import matplotlib.pyplot as plt 

#téléchargement fichier cif du matériau NaInS2
with MPRester("inU5O5NjJKENIfaZf6EO") as m:
    structure = m.get_structure_by_material_id("mp-20289")

#Création de l'objet calculator du modèle de diffraction
xrdcalc = XRDCalculator(wavelength=1.5406)
patt = xrdcalc.get_pattern(structure)

#Les 3 premiers pics du diffractogramme pour un rayonnement X incident CuKα de longueur d’onde λ=1.54060 Å
print('-------------------------------------------------')
print("Premier pic de diffraction du diffractogramme:\n")
print("Angle 2\u03B8 [°] :", patt.x[0])
print("Intensité du pic [%] :", patt.y[0])
print("Indices hkl :", patt.hkls[0])
print('\n')

print('-------------------------------------------------')
print("Deuxième pic de diffraction du diffractogramme :\n")
print("Angle 2\u03B8 [°] :", patt.x[1])
print("Intensité du pic [%] :", patt.y[1])
print("Indices hkl :", patt.hkls[1])
print('\n')

print('-------------------------------------------------')
print("Troisième pic de diffraction du diffractogramme :\n")
print("Angle 2\u03B8 [°] :", patt.x[2])
print("Intensité du pic [%] :", patt.y[2])
print("Indices hkl :", patt.hkls[2])
print('\n')

#Graphe du diffractogramme par l'objet XRDCalculator
print("Graphe du motif de diffraction de la structure pour un rayonnement X incident CuKα de longueur d’onde λ=1.54060 Å sur lequel on peut observer les 3 premiers pics:")
xrdcalc.show_plot(structure,annotate_peaks=True)
```
